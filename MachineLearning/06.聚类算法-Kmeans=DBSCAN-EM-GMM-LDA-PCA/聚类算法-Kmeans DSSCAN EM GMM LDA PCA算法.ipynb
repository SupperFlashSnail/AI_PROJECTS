## K-means聚类算法
```
1. 选取K个样本作为簇中心，并计算所有样本与这个K个簇中心的距离，并将数据划分为K簇
2. 求每一个簇的中心，然后以这个中心坐标更新簇中心
3. 使用更新后的簇中心重新计算样本数据距离，并将数据划分为K簇
4. 重复23操作
```
```
kmeans优缺点：
优点：原理简单，聚类效果优
缺点：缺点比较明显，非常依赖首次聚类簇中心的选取，对噪音比较敏感
```

## DBSCAN算法
```
基于密度的聚类方法：
核心对象：半径为R的领域内样本点个数不少于阈值
直接密度可达：一个核心对象领域内的点相对于核心点而言是直接密度可达
密度可达：密度可达包括直接密度可达和间接密度可达，间接密度可达是两个点对应核心点是密度可达的关系，那么这两个点是间接密度可达

DBSCAN聚类过程：
1. 标记所有点为未访问状态
2. 任意选取一个点，并将其状态改为访问状态，如果该点半径为R的领域内的样本点个数大于阈值，则创建一个新的簇C，该点为核心点
3. 将核心点半径为R的领域内的点归类于簇C，并判断这些点是否是核心点，如果是核心点，则再次向四周蔓延，指导当前簇找不到新的核心点为止。然后从剩余的未被访问的点中任意选取一个点作为下一个簇的核心点。

```

## EM算法
```
EM算法：一种含有隐变量的概率模型的最大似然估计方法。EM算法是在已知数据分布类型 和 无标签样本数据 情况下将数据分为K类。

对于EM算法，其默认的是 对于所有聚类而言，其样本数据的分布类型相同，但是其分布参数不同，进而可以通过最大似然估计的方式，求出每一个聚类数据的分布参数。

下图以正态分布为例：
1. 假设当前目标是将当前数据分为3类，则需要初始化三个分布参数，即三组均值和方差，然后遍历所有样本数据，把每一个样本数据分别代入三组参数，并求出其属于某一个簇的概率值，通过比较，将对应的样本点归类于概率最大的簇

2. 所有样本最终分为三簇，然后通过三簇的数据，计算出每一簇数据的均值和方差

3. 使用计算后的均值和方差更新初始的分布参数

4. 然后重复 23操作

```

## PCA聚类算法
```
PCA 主成分分析算法：提取最优价值的信息，通过坐标系的转变，使数据映射更加分散，进而便于聚类。【其操作的特征数据】
坐标变幻的目标：
1.前提：两个特征向量之间的协方差越小越好[特征之间线性不相关]，默认等于0
2.实现目标：特征数据方差越达越好[类似于熵值，特征数据熵值越大，分布越分散，越容易根据特征数据进行划分]

输入：训练样本集 D=x(1),x(2),...,x(m) ,低维空间维数 d′ ;
过程：.
1：对所有样本进行中心化（去均值操作）： x(i)j←x(i)j−1m∑mi=1x(i)j ;
2：计算样本的协方差矩阵 XXT ;
3：对协方差矩阵 XXT 做特征值分解 ;
4：取最大的 d′ 个特征值所对应的特征向量 w1,w2,...,wd′
5：将原样本矩阵与投影矩阵相乘： X⋅W 即为降维后数据集 X′ 。其中 X 为 m×n 维， W=[w1,w2,...,wd′] 为 n×d′ 维。
5：输出：降维后的数据集 X′
```

## LDA
```
[有监督学习]
LDA 线性判别分析：
LDA和PCA在计算操作上很相似，都是通过坐标系的变换使得聚类数据方便，其前提条件也是坐标变换后，特征之间线性不相关。和LDA 不同的是，PCA是无监督学习，其求解目的是通过坐标系变换使特征数据更加分散，即特征数据方差最大；而LDA是有监督学习，其求解的目的是通过坐标变换使标签中同一类别的数据更加聚合，而不同类别的数据更加分散，即对于对于同一类别的数据方差最小，不同类别数据的方差最大
```

````python
import pandas as pd
%matplotlib inline
data = pd.read_csv ('Fremont.csv', index_col='Date', parse_dates=True)
print(data.head())
print(data.plot())
print(data.resample('w').sum().plot())

data.columns =['West', 'East']
data ['Total'] =data['West']+data['East']
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)

X = pivoted.fillna(0).T.values
X.shape
````

````python
import matplotlib.pyplot as plt
from  sklearn.decomposition import PCA
X2 = PCA(2).fit_transform(X)
print(X2.shape)
plt.scatter(X2[:,0],X2[:,1])
````

````python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=800, centers=4, random_state=11)
plt.scatter(X[:, 0], X[:, 1])
````

````python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
````

````python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4).fit(X)
labels = gmm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
````
